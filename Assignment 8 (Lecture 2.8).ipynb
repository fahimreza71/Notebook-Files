1. Write python codes that shows the arithmatic operations using function.

```python
def sum(*args):
  ans = 0
  for x in args:
    ans+=x
  return ans

print(sum(5,5))
```

2. Find the largest number from three numbers using function.

```python
def max(*args):
  ans=0
  for x in args:
    if(ans<x):
      ans=x

  return ans

print(max(34,56,1))
```
