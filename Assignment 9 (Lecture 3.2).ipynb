10. Write a code that show all the numpy array generation procedure.

```python
# Generate a 3x3 array of zeros
zeros_array = np.zeros((3, 3))
print("Zeros array:\n", zeros_array)

# Generate a 3x3 array of ones
ones_array = np.ones((3, 3))
print("\nOnes array:\n", ones_array)

# Generate a 3x3 array of random numbers
random_array = np.random.rand(3, 3)
print("\nRandom array:\n", random_array)

# Generate a 3x3 array of random integers
random_int_array = np.random.randint(low=0, high=10, size=(3, 3))
print("\nRandom integer array:\n", random_int_array)

# Generate a 3x3 array of evenly spaced values
evenly_spaced_array = np.linspace(start=0, stop=10, num=9).reshape(3, 3)
print("\nEvenly spaced array:\n", evenly_spaced_array)

# Generate a 3x3 identity matrix
identity_matrix = np.eye(3)
print("\nIdentity matrix:\n", identity_matrix)
```
