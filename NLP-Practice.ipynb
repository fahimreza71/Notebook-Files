#Spam Classification

```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/Datasets/SMSSpamCollection',sep='\t',names=['label','message'])
df
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
```

**Classification Using Steamming --> BoW --> MultinomialNB**

```python
# Using Steamming
corpus = []
for i in range(len(df)):
  review = re.sub('[^a-zA-Z0-9]', ' ', df['message'][i])
  review = review.lower()
  review = review.split()
  review = [stemmer.stem(word) for word in review if not word in set(stopwords.words('english'))]
  review = ' '.join(review)
  corpus.append(review)
```

```python
corpus
```

```python
# Converting Word --> Vector (BoW)
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500)
X = cv.fit_transform(corpus).toarray()
```

```python
type(X)
```

```python
y = pd.get_dummies(df['label'])
y = y.iloc[:,1].values
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB().fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_m = confusion_matrix(y_test, y_pred)
confusion_m
```

```python
from sklearn.metrics import accuracy_score, classification_report
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
print(classification_report(y_test, y_pred))
```

**Classification Using Lemmatization --> TF-IDF --> RandomForestClassifier**

```python
# Using Lemmatization
corpus = []
for i in range(len(df)):
  review = re.sub('[^a-zA-Z0-9]', ' ', df['message'][i])
  review = review.lower()
  review = review.split()
  review = [lemmatizer.lemmatize(word) for word in review if not word in set(stopwords.words('english'))]
  review = ' '.join(review)
  corpus.append(review)
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
cv = TfidfVectorizer(max_features=2500, binary=True, ngram_range=(1, 3))
X = cv.fit_transform(corpus).toarray()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier().fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, accuracy_score
score = accuracy_score(y_test, y_pred)
score
```

```python
print(classification_report(y_test, y_pred))
```

```python
# Using Lemmatization
corpus = []
for i in range(len(df)):
  review = re.sub('[^a-zA-Z0-9]', ' ', df['message'][i])
  review = review.lower()
  review = review.split()
  review = [lemmatizer.lemmatize(word) for word in review if not word in set(stopwords.words('english'))]
  review = ' '.join(review)
  corpus.append(review)
```

```python
[[i,j,k] for i,j,k in zip(list(map(len,corpus)),corpus, df['message']) if i<1]
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
nltk.download('punkt_tab')
```

```python
words = []
for sentence in corpus:
  sent_token = sent_tokenize(sentence)
  for sent in sent_token:
    words.append(simple_preprocess(sent))
```

```python
words[0]
```

```python
# Train Custom Word2Vec from Scratch
import gensim
model = gensim.models.Word2Vec(words, window=5, min_count=2)
```

```python
model.wv.index_to_key
```

```python
model.wv.most_similar('coffee')
```

```python
# AvgWord2Vec from Custom Word2Vec
import numpy as np
def avg_word2vec(doc):
  vectors = [model.wv[word] for word in doc if word in model.wv.index_to_key]
  # return np.mean(vectors, axis=0)
  if len(vectors) > 0:
    return np.mean(vectors, axis=0)
  else:
    return np.zeros(model.vector_size)
```

```python
X = []
for doc in words:
  X.append(avg_word2vec(doc))
```

```python
type(X)
```

```python
X = np.array(X)
# X = [np.array(x) for x in X]
```

```python
print(type(X))
print(X.shape)
```

```python
y = df[list(map(lambda x: len(x)>0 ,corpus))]
y = pd.get_dummies(df['label'])
y = y.iloc[:,1].values
```

```python
y.shape
```

```python
print(f"Length of X: {len(X)}")
print(f"Length of y: {len(y)}")
```

```python
filtered_X = []
filtered_y = []

for i, x in enumerate(X):
    if len(x) != 0:
        filtered_X.append(x)
        filtered_y.append(y[i])

X = np.array(filtered_X)
y = np.array(filtered_y)

print(f"Filtered X shape: {X.shape}")
print(f"Filtered y shape: {y.shape}")
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from xgboost import XGBClassifier
model = XGBClassifier().fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, accuracy_score
y_pred = model.predict(X_test)
score = accuracy_score(y_test, y_pred)
score
```

```python
print(classification_report(y_test, y_pred))
```
