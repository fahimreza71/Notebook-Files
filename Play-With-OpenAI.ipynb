```python
import pandas as pd
import json

# Load the CSV file
df = pd.read_csv("text-dataset.csv")

# Keep only necessary columns
df = df[['text', 'label']].dropna()

# Convert to JSONL format
jsonl_path = "text-dataset.jsonl"
with open(jsonl_path, "w") as file:
    for _, row in df.iterrows():
        json_obj = {"prompt": row["text"], "completion": str(row["label"])}
        file.write(json.dumps(json_obj) + "\n")

print(f"JSONL file saved as {jsonl_path}")
```

```python
import json

file_path = 'text-dataset_prepared_train.jsonl'

examples = []
with open(file_path, 'r') as file:
    for line in file:
        data = json.loads(line)
        examples.append(data)

for example in examples:
    print(f"Prompt: {example['prompt']}")
    print(f"Completion: {example['completion']}")
    print("-" * 50)
```

```python
prompt = ""
for example in examples:
    prompt += f"{example['prompt']}\n{example['completion']}\n\n"

test_prompt = "energy_100g:0.0,fat_100g:0.0,carbohydrates_100g:0.0,sugars_100g:0.0,proteins_100g:0.0,salt_100g:2.38252,sodium_100g:0.938"
prompt += f"{test_prompt}\nOutput:"

print(prompt)
```

```python
messages = [
    {"role": "system", "content": "You are a nutritional assistant. Based on the input food data, return the health value."},
    {"role": "user", "content": "energy_100g:192.0,fat_100g:1.08,saturated-fat_100g:0.62,trans-fat_100g:0.0,cholesterol_100g:0.006,carbohydrates_100g:5.23,sugars_100g:5.23,fiber_100g:0.0,proteins_100g:3.38,salt_100g:0.13208,sodium_100g:0.052,vitamin-a_100g:6.93e-05,vitamin-c_100g:0.0,calcium_100g:0.123,iron_100g:0.0 ->"}
]
```

```python
response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=messages,
    max_tokens=100,
    temperature=0.7
)

# Print the output
# print("Model's output:", response.choices[0].message['content'].strip())
response.choices[0].message.content
```
